# Counting Words

```python
import swifter
from nltk.tokenize import TweetTokenizer
from nltk.util import ngrams
from tqdm import tqdm
```

```python
text_df = tweets_df[['tweet_text', 'tokens']].copy()
```

```python
def get_ngrams(text, n):
    n_grams = ngrams(TweetTokenizer().tokenize(text), n)
    return [' '.join(grams) for grams in n_grams]
def get_tokens(loc_s):
    return get_ngrams(loc_s, 1) + get_ngrams(loc_s, 2)
```

```python
#tqdm.pandas()
#text_df['tokens'] = text_df['tweet_text'].progress_apply(get_tokens)
text_df.head()
```

```python
tweets_df['tokens'] = text_df['tokens']
tweets_df.to_parquet(parquet_path, compression='gzip')
```

```python
with open('../public_data/body_vocab.txt', 'r') as f:
    body_words_list = [word.strip('\n"') for word in f.readlines()]
body_words_list
# Won't add up to total count of tweets
# Note "her face" appears but not "his face"
from collections import Counter
c = Counter()
for token_list in tqdm(text_df['tokens'].values, total=len(text_df['tokens'])):
    for token in body_words_list:
        if token in token_list:
            c[token] += 1
c
```

```python
words_df = pd.DataFrame.from_dict(c, orient='index', columns=['tweet_count'])
words_df['token'] = words_df.index
words_df.head()
```

```python
fig, ax = plt.subplots(1,1,figsize=(7,5.5))
ax = sns.barplot(y='token', x='tweet_count', data=words_df.sort_values('tweet_count', ascending=False), color='black')
ax.xaxis.grid()
ax.set_ylabel('Token')
ax.set_xlabel('Tweet Count')
plt.savefig(save_dir + 'token_tweet_counts.pdf', dpi=300)
```

```python
small_df = tweets_df[tweets_df['tweet_text'].str.contains('body')]
small_df.head()
```

```python
def get_word_emot_score(word_row, score_df):
    small_df = score_df[score_df['tweet_text'].str.contains(word_row['token'])]
    return small_df['joy'].mean()
```

```python
words_df['joy_score'] = words_df.progress_apply(get_word_emot_score, args=(tweets_df[['tweet_text','joy']],), axis=1)
words_df.head()
```
